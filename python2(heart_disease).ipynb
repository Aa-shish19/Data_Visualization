```python
sns.histplot(df['age'], kde=True, color='grey')

plt.axvline(df['age'].mean(), color='red', linestyle='--')
plt.axvline(df['age'].median(), color='green', linestyle='-')

plt.title('Distribution of Age')
plt.legend(['Median', 'Mean'])
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.title('Count of Sex')
plt.show()
```

```python
sns.countplot(x='target', data=df, palette='Set2')
plt.title('Distribution of Target Variable')
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
sns.countplot(x='sex', hue='target', palette='Set1', data=df)
plt.title('Count of Sex by Target')
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
sns.boxplot(x='target', y='age', data=df, palette='Set2')
plt.title('Distribution of Age by Target')
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
sns.boxplot(x='target', y='thalach', data=df, palette='Set2')
plt.title('Distribution of Maximum Heart Rate by Target')
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
sns.boxplot(x='target', y='chol', data=df, palette='Set2')
plt.title('Distribution of Cholesterol by Target')
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
num = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca', 'thal', 'target', 'cp', 'sex', 'fbs', 'restecg', 'exang', 'slope']
sns.heatmap(df[num].corr(), cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
crosstab = pd.crosstab(df['cp'], df['target'], normalize='index')*100
crosstab.plot(kind='bar', stacked=True, color=['lightblue', 'salmon'])
plt.title('Chest Pain Type by Target')
plt.xlabel('Chest Pain Type')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```
