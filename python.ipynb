```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
missing = df.isnull().sum()
print(missing[missing > 0].sort_values(ascending=False))
```

```python
sns.histplot(df['Age'], kde=True, color='blue')#kde=True adds a kernel density estimate to the plot, which gives a smoothed curve representing the distribution of the data. The color parameter sets the color of the histogram bars and the KDE line.

plt.axvline(df['Age'].mean(), color='red', linestyle='--', label='Mean') #This line adds a vertical line to the plot at the position of the mean age. The color is set to red, the linestyle is dashed, and a label 'Mean' is provided for the legend.
plt.axvline(df['Age'].median(), color='green', linestyle='-', label='Median') #This line adds another vertical line at the position of the median age. The color is green, the linestyle is solid, and it is labeled 'Median' for the legend.

plt.legend() #This adds a legend to the plot, which helps identify the mean and median lines based on their colors and styles.
plt.title('Distribution of Age on the Titanic')

plt.show()
```

```python
df['Survived'].value_counts().plot(kind='bar', color=['red', 'green'])
plt.xticks([0,1],['Died', 'Survived'], rotation=0)
plt.title('Survival Count on the Titanic')
plt.show()
```

```python
ct = pd.crosstab(df['Sex'], df['Survived'], normalize='index')*100 
ct.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.xticks(rotation=0)
plt.ylabel('Percentage')
plt.title('Survival Rate by Sex on the Titanic')
plt.legend(title='Survived', labels=['Died', 'Survived'])
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.7, palette={0:'red', 1:'green'})
plt.title('Age vs Fare by Survival on the Titanic')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df, palette='Set2')
plt.title('Age Distribution by Passenger Class on the Titanic')
plt.show()
```

```python
num = df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'Survived']]
sns.heatmap(num.corr(), annot=True, fmt='.2f', cmap='coolwarm',center=0)
plt.title('Correlation Heatmap of Numerical Features on the Titanic')
plt.show()
```

```python
sex_survived = (df.groupby('Sex')['Survived'].mean()*100).round(1)
print(sex_survived)
```

```python
pass_survived = (df.groupby('Pclass')['Survived'].mean()*100).round(1)
print(pass_survived)
```

```python
df['Child'] = df['Age']<16
child_survived = (df.groupby('Child')['Survived'].mean()*100).round(1)
print(child_survived)
```

```python
child_survived.plot(kind='bar', color=['red', 'green'])
plt.xticks([0,1],['Adult', 'Child'], rotation=0)
plt.ylabel('Survival Rate (%)')
plt.title('Survival Rate by Age Group on the Titanic')
plt.show()
```

```python
likely_survived = (df.groupby('PassengerId')['Survived'].mean()*100).round(1)
print(likely_survived)
```

```python
sns.scatterplot(x='PassengerId', y='Age', hue='Survived', data=df, alpha=0.7, palette=['red', 'green'])
plt.title('Passenger ID vs Age by Survival on the Titanic')
plt.show()
```
